<h1 style="text-align:center;font-size:02.2vw;">SSL4EO-S12 dataset</h1>

In this use case we are going to develop a dataset consisting of approximately 10M S1-S2 image pairs for pre-training custom architectures with multi-sensory input data. This will increase on one order of magnitude with respect to the biggest public EO dataset available today for this task. In order to do this, the main tools that will be tested are: Define worldwide geographically distributed AOIs and massively download S1 and S2 imagery (as close in dates as possible) with temporal dimension. The dataset will be useful for pre-training neural networks in an unsupervised way in order to provide better initialization for downstream tasks with increased data efficiency and enable the exploration of custom EO architectures with good level of performance.

# Dataset

The `SSL4EO-S12` dataset is a large-scale multimodal multitemporal dataset for unsupervised/self-supervised pre-training in Earth observation. The dataset consists of unlabeled patch triplets (Sentinel-1 dual-pol SAR, Sentinel-2 top-of-atmosphere multispectral, Sentinel-2 surface reflectance multispectral) from 251079 locations across the globe, each patch covering 2640mx2640m and including four seasonal time stamps.

Original publication and info at: https://mediatum.ub.tum.de/1660427?v=1

Github repository: https://github.com/zhu-xlab/SSL4EO-S12

# Check available data from AIRBUS

With the `geometry` column, we can create the `dict` needed by the EO-TDL library to check for available images, which is a `dict` with format `location_id`: { `bounding_box` , `time_interval` }.

```python
from eotdl.tools import generate_location_payload, format_product_location_payload

bbox_date_by_location = generate_location_payload(images_by_location, path='preprocess')

print(bbox_date_by_location)
```

Before using the Airbus API, we need to authenticate and get an API KEY, in order to generate an access token.

```python
from eotdl.access.airbus.utils import get_airbus_access_token
from eotdl.access.airbus.client import AirbusClient

airbus_api_key = getenv("AIRBUS_API_KEY")   # Put your Airbus API key here
access_token = get_airbus_access_token(airbus_api_key)
```

```python
airbus_client = AirbusClient(access_token)
```

Let's search for images!

If you have already searched them, they have been stored in a `.json` file, so you don't need to search them again.

```python
if exists('preprocess/airbus_images_response.json'):
    with open('preprocess/airbus_images_response.json', 'r') as f:
        images_response = json.load(f)
```

If you have not, search for them!

```python
images_response = airbus_client.search_images_close_in_time(bbox_date_by_location, path='preprocess', max_days=7)
```

Now that we have a response with the available SPOT images for every scene, we need to reformat the payload dict and add the product `id` of every Airbus image to the images payload dict that we have generated before, as we are going to need it to track the location of every image.

```python
images_payload = format_product_location_payload(bbox_date_by_location, images_response, all_info=False)
images_payload
```

Now, we can split the responses between scenes with Airbus images and scenes without them.

```python
scenes_with_airbus_images, scenes_without_airbus_images = airbus_client.split_product_payload(images_payload)
print(scenes_with_airbus_images)
print(scenes_without_airbus_images)
```

```python
# Save the scenes with Airbus images as json
with open('preprocess/scenes_with_airbus_images.json', 'w') as f:
    json.dump(scenes_with_airbus_images, f)
```

```python
print(f'Number of scenes with Airbus images: {len(scenes_with_airbus_images)}')
print(f'Number of scenes without Airbus images: {len(scenes_without_airbus_images)}')
```

Before ordering the images we need, we are going to take a look to their prices.

```python
for location_id, location_data in scenes_with_airbus_images.items():
    product_id = location_data['image']
    if id:
        bbox = location_data['bounding_box']
        price = airbus_client.get_product_price(product_id, bbox)
        print(location_id, price['price'])
```
